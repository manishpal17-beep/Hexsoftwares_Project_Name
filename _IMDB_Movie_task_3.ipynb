Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
data.columns
```

```python
def rating(rating):
    if rating>=8.0:
        return "Excellent"
    elif rating>=7.0:
        return "Good"
    else :
        return "Average"
```

```python
data['rating_cat']=data['Rating'].apply(rating)
```
